# Results and Comparison of CNN models

```python
import pandas 
import matplotlib
import matplotlib.pyplot as plt
```

```python
lines = pandas.read_csv('./history/results.csv', index_col=0)
```

```python
lines.head()
```

## Test accuracy

```python
x = list(lines.index)
means = list(lines['test acc'])
stds = list(lines['test acc std'])

plt.figure(num=1, figsize=(16, 8))
plt.rc('xtick', labelsize=16) 
plt.rc('ytick', labelsize=16) 
plt.errorbar(x, means, stds, fmt='o', color='white', ecolor='black', elinewidth=15, capsize=10);
```

## Test loss

```python
x = list(lines.index)
means = list(lines['test loss'])
stds = list(lines['test loss std'])

plt.figure(num=1, figsize=(16, 8))
plt.rc('xtick', labelsize=16) 
plt.rc('ytick', labelsize=16) 
plt.errorbar(x, means, stds, fmt='o', color='white', ecolor='black', elinewidth=15, capsize=10);
```

## Train-Test accuracy difference

```python
x = list(lines.index)
means = list(lines['train acc'] - lines['test acc'])
stds = [0]*4

plt.figure(num=1, figsize=(16, 8))
plt.rc('xtick', labelsize=16) 
plt.rc('ytick', labelsize=16) 
plt.errorbar(x, means, stds, fmt='o', color='red', ecolor='black', elinewidth=15, capsize=10);
```

## Train-Test loss difference

```python
x = list(lines.index)
means = list(lines['test loss'] - lines['train loss'])
stds = [0]*4

plt.figure(num=1, figsize=(16, 8))
plt.rc('xtick', labelsize=16) 
plt.rc('ytick', labelsize=16) 
plt.errorbar(x, means, stds, fmt='o', color='red', ecolor='black', elinewidth=15, capsize=10);
```
